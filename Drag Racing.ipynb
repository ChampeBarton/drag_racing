```python
# import libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
url_list = ["https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=1&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=2&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=3&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=12&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=28&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=4&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=5&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=6&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=7&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=8&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=160&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=11&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=10&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=30&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=34&rn=3",
           "https://www.nhraeventreg.com/ListEventCatgsiFrame.asp?ey=2018&ec=GF1&cc=114&rn=3"]
```

```python
drivers = []
```

```python
for i in range(0, 16):
    page = urlopen(url_list[i])
    soup = BeautifulSoup(page, "html.parser") # parse the html using beautiful soup and store in variable `soup`
    divs = soup.findAll("td")
    
    for j in range(1, (len(divs)//7)):
        drivers.append(divs[j*7+2].text.strip())
```

```python
drivers
```

```python
# open a csv file with append, so old data will not be erased
with open('DragDrivers.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(["name"]) # write csv file headers
    for i in range (0, len(drivers)):
        writer.writerow([drivers[i]])
```
